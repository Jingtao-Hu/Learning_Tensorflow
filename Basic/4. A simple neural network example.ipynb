```python
import tensorflow as tf

from numpy.random import RandomState

# Define batch size
batch_size = 8

# Define variables 'w1' and 'w2' and set random_seed for keeping every run result same
w1 = tf.Variable(tf.random_normal([2,3],stddev=1,seed=1))
w2 = tf.Variable(tf.truncated_normal([3,1],stddev=1,seed=1))

# Define the input and ground truth
x = tf.placeholder(tf.float32,shape=[None,2],name='x_input')
y_ = tf.placeholder(tf.float32,shape=[None,1],name='y_input')

# Define the forward-propagation of neural network
a = tf.matmul(x,w1)
y = tf.matmul(a,w2)

# Define loss function and back-propagation
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=y,labels=y_))
optimizer = tf.train.AdamOptimizer(learning_rate=0.001)
train_step = optimizer.minimize(cross_entropy)

# Create dataset
rdm = RandomState(1)
dataset_size = 128
X = rdm.rand(dataset_size,2) # shape is (128,2)
Y = [[int(x1+x2 < 1)] for (x1,x2) in X]

# Create Session and run the tensorflow program
with tf.Session() as sess:
    init = tf.global_variables_initializer()
    sess.run(init)
    print("The parameters of pre-training:")
    print(sess.run(w1))
    print(sess.run(w2))
    
    steps = 5000
    for i in range(steps):
        # Pick up the number of batch_size data to train
        start = (i*batch_size) % dataset_size
        end = min(start+batch_size, dataset_size)
        
        # Training and update parameters
        sess.run(train_step,feed_dict={x:X[start:end],y_:Y[start:end]})
        
        if i % 1000 == 0:
            # output the cross entropy per 1000 times run
            total_cross_entropy = sess.run(cross_entropy,feed_dict={x:X,y_:Y})
            print("After %d training steps, cross entropy on all data is %f" %
                 (i, total_cross_entropy))

    print("The parameters of post-training:")
    print(sess.run(w1))
    print(sess.run(w2))
```
